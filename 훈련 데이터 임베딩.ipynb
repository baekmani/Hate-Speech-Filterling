```python
import pandas as pd
import numpy as np
from Data_preprocessing import data_split
from gensim.models import FastText
```

```python
data = pd.read_csv("hate_train.csv")
data.drop('Unnamed: 0', axis= 1, inplace = True)
data
```

```python
# 훈련데이터 초성, 중성, 종성 분리
data['comments'] = data['comments'].apply(lambda x: data_split(x))
```

```python
data.head()
```

```python
# fasttext 임베딩 용 모델 불러오기 - 다른 임베딩은 적용 X
from gensim.models import FastText
global embedding_model
embedding_model = FastText.load("./festtext.model")

def embedding(x):
    return embedding_model.wv[x]

data['comments'] = data['comments'].apply(embedding)
```
